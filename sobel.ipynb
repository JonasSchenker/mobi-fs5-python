**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def sobel_h(img):
    """Horizontal Sobel filter (detects vertical edges)"""
    kernel_h = np.array([[-1, 0, 1],
                         [-2, 0, 2],
                         [-1, 0, 1]])
    
    h, w = img.shape
    result = np.zeros((h - 2, w - 2)) #Because the kernel is 3x3, the output image is smaller by 2 pixels in each dimension.
    
    for i in range(h - 2):
        for j in range(w - 2):
            result[i, j] = np.sum(img[i:i+3, j:j+3] * kernel_h)
    
    return result
    
    
def sobel_v(img):
    """Vertical Sobel filter (detects horizontal edges)"""
    kernel_v = np.array([[-1, -2, -1],
                         [0, 0, 0],
                         [1, 2, 1]])
    
    h, w = img.shape
    result = np.zeros((h - 2, w - 2))
    
    for i in range(h - 2):
        for j in range(w - 2):
            result[i, j] = np.sum(img[i:i+3, j:j+3] * kernel_v)
    
    return result
```

**[Task 2.3.]()** Test your implementations from above:

**[Task 2.4.]()** Compute the image gradient magnitude:
