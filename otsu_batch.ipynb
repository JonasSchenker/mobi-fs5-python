**[Task 2.1.]()** Implement Otsu thresholding as a re-usable function:

```python
def otsu_thresholding(imgf):
    ''' Input: imgf - grayscale image with pixel values in [0,1]'''
    img = (imgf * 255).round().astype(np.uint8)
    sigma_b_squared_max = 0
    threshold = 0

    # compute histogram
    h = []
    for i in range(0,256):
        bool_img = img == i
        pixelcount = bool_img.sum()
        h.append(pixelcount)

    h = np.array(h)

    #total number of pixels
    for T in range(1, 256):
        n1 = np.sum(h[0:T])
        n2 = np.sum(h[T:256])
        if n1 == 0 or n2 == 0: #skip to next value of T if one class is empty, cannot compute variance
            continue    
        # compute class means
        mean1 = np.sum([i * h[i] for i in range(0, T)]) / n1 
        mean2 = np.sum([i * h[i] for i in range(T, 256)]) / n2

        # compute the between class variance
        sigma_b_squared = n1 * n2 * (mean1 - mean2) ** 2
    
        # find maximum sigma_b_squared
        if sigma_b_squared > sigma_b_squared_max :
            sigma_b_squared_max = sigma_b_squared
            #update threshold to the current T with maximum between class variance
            threshold = T 

    print('threshold =' + str(threshold))

    # Perform otsu thresholding
    binary_img = img >= threshold  
   
    seg = binary_img.astype(np.uint8) * 255 # convert boolean to uint8 image
    
    return seg
```

**[Task 2.2.]()** Implement a re-usable function to compute the Dice coefficient:

```python
def compute_dice(bin1, bin2):
    assert bin1.dtype == bool
    assert bin2.dtype == bool

    dice_score = 2 * np.sum(bin1 * bin2) / (np.sum(bin1) + np.sum(bin2))
    
    return dice_score
```
